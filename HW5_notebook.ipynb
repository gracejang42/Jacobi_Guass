```python
import numpy as np
import math
import scipy
import scipy.linalg
import time
#Grace Horton
#Homework 5
#Collaborators: None

file = open("5mat3.txt")
A = np.loadtxt(file, delimiter=",")
size = np.size(A)
rows = len(A)
cols = len(A[0])
b = np.random.randint(100,size=(rows,1))
n = rows
tol = 1e-10
max_i = 100000


def jacobi(A,b,max_iter):
    x = np.zeros_like(b, dtype=np.double)
    #rel_diff = np.zeros_like(b, dtype=np.double) 
    for COUNT in range(max_iter):
        x_new = x.copy()
        for i in range(A.shape[0]):
            # finding C_i
            s1 = np.dot(A[i, :i], x[:i])
            s2 = np.dot(A[i, i+1:], x[i+1:])
            # Dx_i = C_i for x_i+1
            x_new[i] = (b[i] - s1 - s2)/ A[i,i]
            # Calculate the Relative Error
            #rel_diff[i] = np.linalg.norm(x_new - x, ord=np.inf)/ np.linalg.norm(x, ord=np.inf) 
        #Stop condition
        if np.allclose(x, x_new, atol=1e-10, rtol=0):
            break
        x = x_new
    return x # rel_diff


def gauss_seidel(A, b, tol, max_iter):    
    x = np.zeros_like(b, dtype=np.double) 
    #rel_err = np.zeros_like(b, dtype=np.double) 
    #Iterate
    for k in range(max_iter):
        x_old  = x.copy()
        #Loop over rows
        for i in range(A.shape[0]):
            x[i] = (b[i] - np.dot(A[i,:i], x[:i]) - np.dot(A[i,(i+1):], x_old[(i+1):])) / A[i ,i]
            #rel_err[i] = np.linalg.norm(x - x_old, ord=np.inf) / np.linalg.norm(x, ord=np.inf)
        #Stop condition 
        if np.linalg.norm(x - x_old, ord=np.inf) / np.linalg.norm(x, ord=np.inf) < tol:
            break
            
    return x # rel_err

tic = time.perf_counter()
truth = np.linalg.solve(A,b)
toc = time.perf_counter()
print(f"TRUTH completed in {toc - tic:0.4f} seconds")
tic2 = time.perf_counter()
j_sol = jacobi(A,b,max_i)
toc2 = time.perf_counter()
print(f"Jacobi completed in {toc2 - tic2:0.4f} seconds")
tic3 = time.perf_counter()
g_sol = gauss_seidel(A,b,tol,max_i)
toc3 = time.perf_counter()
print(f"Guass-Seidel completed in {toc3 - tic3:0.4f} seconds")

# Reporting Test Values

loopyboi = 5
tmp = np.empty(loopyboi, dtype=np.double) 
tmp2 = np.empty(loopyboi, dtype=np.double) 
tmp3 = np.empty(loopyboi, dtype=np.double) 
#tmp4 = np.empty(loopyboi, dtype=np.double) 
#tmp5 = np.empty(loopyboi, dtype=np.double) 
for i in range(0,loopyboi):
    tmp[i] = truth[i]
    tmp2[i] = j_sol[i]
    tmp3[i] = g_sol[i]
    #tmp4[i] = diff[i]
    #tmp5[i] = g_err[i]
theFather = tmp
theSon = tmp2
theHolyGhost = tmp3
#j_sin = tmp4
#g_sin = tmp5


temp = np.empty(loopyboi, dtype=np.double)
temp2 = np.empty(loopyboi, dtype=np.double)
for i in range(0,loopyboi):
    temp[i] = np.linalg.norm(theFather - theSon, ord=np.inf) / np.linalg.norm(theFather, ord=np.inf)
    temp2[i] = np.linalg.norm(theFather - theHolyGhost, ord=np.inf) / np.linalg.norm(theFather, ord=np.inf)

blaspheme = temp
heretic = temp2

#print('==================== Relative Errors: First 5 Entries ===================')
#print('Jacobi Method:', j_sin)
#print('-------------------------------------------------------------------------')
#print('Guass-Seidel Method:', g_sin)
#print('.........')
print('========== Relative Errors Compared to Truth: First 5 Entries: ==========')
print('Jacobi Method:', blaspheme)
print('-------------------------------------------------------------------------')
print('Guass-Seidel Method:', heretic)
print('-------------------------------------------------------------------------')
print('.........')

print('=============================================')
print('Solution Vectors, first 5 entires:')
print('---------------------------------------------')
print('Truth Solution Vector       :', theFather)
print('---------------------------------------------')
print('Jacobi Solution Vector      :', theSon)
print('---------------------------------------------')
print('Guass-Seidel Solution Vector:', theHolyGhost)
print('---------------------------------------------')

#print("Jacobi ........Jacobi error............ Guass-Seidel... ... ...  Guass-Seidel Error")
#print("=========================================================================")
#res = "\n".join("{} ... ... {} ... ... {} ... ... {}".format(q, r, x, y) for q,r,x,y in zip(j_sol, diff, g_sol, g_err))
#print(res)
```
